# Fibonacci series problem [Approache-2 :Memorization (top-down approach) ] complexity:
### here time taken for retrieving hash-values from hash-table will be constant O(1)
### whereas time taken for finding fibonacci-value of number-n will be of order n [O(n)] as with the help of hash-table maximum n-times function called:


# Fibonacci series problem [Approache-2 :Memorization (top-down approach) ] code-implementation:




```python
def fibonacci_memorization(n, Hash_t={0:0, 1:1}): 
                    # here we intitialized out hashtable with hash-values 1 and 0 for hash-0 & hash-1 respectively (bydefault-values here).
        #checking for function values in hash-table if existedin the hash will return hash-values to respective recursive-fn-calls
        if n in Hash_t:
            return Hash_t[n]
        Hash_t[n] = fibonacci_memorization(n-1, Hash_t) + fibonacci_memorization(n-2, Hash_t)
        return Hash_t[n]  # return the final value of hash for number n 

# Example usage
print(fibonacci_memorization(5))  # Output: 5
print(fibonacci_memorization(10))  # Output: 55
```

```python
#just variation :

#Solution 2
ht={0: 0, 1: 1}
def fibonacci(n):
    if n in ht:
        return ht[n]
    else:
        ht[n] = fibonacci(n - 1) + fibonacci(n - 2)
        return ht[n]
# Example usage
print(fibonacci(5))  # Output: 5
print(fibonacci(10))  # Output: 55
```
