# Climbing-Stairs problem statement:

### why its cosidered to be a application of fibonacci? :

### 'coz if you see ways of climbing 1-stairs it would be 1 only ; for 2-stairs it would be 2 only [these're base-cases certainly]
### but when we go further gretar than 2 like climbing 3 stairs it would be 3 ways ; 
### and climbing 4-stairs [as 4-stairs can be made of 4times 1-stairs(1,1,1,1) or 2 times 2-stairs(2,2) or 1times 2-stairs and 2 times 1-stairs( and can be in this fassion : 2,1,1---1,2,1---1,1,2)  ] : so there're total 5 distinct ways:  

1.  1, 1, 1, 1
2.  1, 1, 2
3.  1, 2, 1
4.  2, 1, 1
5.  2, 2
### if stairs(indexes of 'ways'-array)=[1,2,3,4,5] 
-                                 ways=[1,2,3,5,8]
### so there're 5 ways to climb 4 stairs ; this pattern looks like finonacci-series ; so we're going to use fibonacci-formula to find ways of climbing n-stairs: 
### ways(n)=  ways(n-1) + ways(n-2)

# Climbing-Stairs problem statement test-cases:


```python
#1st-hand code implementation
def climbStairs(n):
    # base-cases will be like:- 
    # a.if 1 stair to climb ; only 1 way to climb which's 1 step at atime
    # b. if 2 stair to climb ; only 2 way to climb which's 1-step(2-times) and 2-step(1-times)
    if (n==1) or (n==2): return n 
    
    #n>=1 ; if stairs(indexes of 'ways'-array)=[1,2,3,4,5] ways=[1,2,3,5,8]   if n=3 ; three ways to climb stairs ; if n=5 (as if there're 4 stairs to climb ; means we have:     2-stairs to climb 2-times(2,2) + 1-stairs to climb 4 times(1,1,1,1) + 1-stairs to climb 2-times and 2 stairs to climb 1-times(1,2,1---1,1,2---2,1,1) = 5
    # so there're 5 ways to climb 4 stairs ; this pattern looks like finonacci-series ; so we're going to use fibonacci-formula to find ways of climbing n-stairs: ways(n)=  ways(n-1) + ways(n-2)
    ways = climbStairs(n-1)+ climbStairs(n-2)
    return ways
    
print(climbStairs(4))  # checking for n=4 ; ways should be 5
print(climbStairs(5))  # checking for n=5 ; ways should be 8
```

# Climbing-Stairs problem statement (fibonacci approach intuition):

### further we can optimized the approach same as we did in fibonacci problem using these further approaches like memoization(top-down) , tabulation(bottom-up) and space optimized tabulation approach.


```python
# prescribed code - implementation: [SPACE_OPTIMIZED tabulation approach; genrally its take place iteratively]:

# this one with with recursive approach:
def climbStairs(n):
    if n<=2: return n
    def helper(first,second,n,count):
        
        next = first + second
        if count==n: return next  # when count reaches n simply returns next 
        return helper(second,next,n,count+1)
    return helper(1,2,n,3)   # starts our prev/first and curr/second with 1 and 2 respectively here   starts from 3   //base condition

print(climbStairs(4))  # checking for n=4 ; ways should be 5
print(climbStairs(5))  # checking for n=5 ; ways should be 8
```
