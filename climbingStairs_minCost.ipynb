# Climbing Stairs min-cost problem:
- Coding Exercise: Minimum Cost Climbing Stairs
- You are given an integer array cost where cost[i] is the cost of ith step on a staircase. Once you pay the cost, you can either climb one or two steps.

[note]: You can either start from the step with index 0, or the step with index 1.

- Return the minimum cost to reach the top of the floor.


Example 1:

Input: cost = [10,20,30]

Output: 20

Explanation: You will start at index 1.
- Pay 20 and climb two steps

```python
# code implementation of recursive approach to solve the climbing stairs min cost problem.

# building a function which takes array of costs of corresponding indexes
def Min_Cost_ClimbingStairs(cost):
    #total steps will equal to the length of cost-array
    n= len(cost)
    def recursive_mincost_stairs(index):    # building here function which will return minimum cost of climbing to the top from particular index
        
        #base-case // if climbing to top done means we're reached beyond the final step  
        if index > (n-1):
            return 0    #then no need to add the cost further; will return zero
        
        #if we're going to take step-1 at a time   //recrusive call by increamenting by 1
        one_step = cost[index] + recursive_mincost_stairs(index+1)   # cost of step is same as ith-number at cost-array
        #if we're going to take step-2 at a time  //recrusive call by increamenting by 2
        two_step = cost[index] + recursive_mincost_stairs(index+2)  # cost of step is same as ith-number at cost-array
        
        return min(one_step,two_step)
    
    #now need to initialize recursive function with intial indexes [either from index-0 or index-1] even from that we're going to return minimum-cost among them 
    return min (recursive_mincost_stairs(0), recursive_mincost_stairs(1))
    
    
#example-usage:
cost = [10,20,30]   # min-cost should be 20 from index-1 and reaching top while climbing 2-step
print(Min_Cost_ClimbingStairs(cost))
         
```

```python
# code implementation of approach memoization for solving the min cost of climbing stairs:

# building a function which takes array of costs of corresponding indexes
def Min_Cost_ClimbingStairs(cost):
    #total steps will equal to the length of cost-array
    n = len(cost)
    #1-d table (array) for maintaining and storing min-cost of reaching to the top from respective index
    min_cost = [-1]*n   # initiaized with -1 at every index
    def memoization_mincost_stairs(index):    # building here function which will return minimum cost of climbing to the top from particular index
        
        #base-case // if climbing to top done means we're reached beyond the final step  
        if index > (n-1):
            return 0    #then no need to add the cost further; will return zero
        
        if min_cost[index] != -1:   # if in min-cost array there's no -1 means min-cost already calculated for that index  
            return min_cost[index]   # simply return min-cost from this index to reaching top
        
        #else do the same old process of calculating min-cost for index :
        
        #if we're going to take step-1 at a time   //recrusive call by increamenting by 1
        one_step = cost[index] + memoization_mincost_stairs(index+1)   # cost of step is same as ith-number at cost-array
        #if we're going to take step-2 at a time  //recrusive call by increamenting by 2
        two_step = cost[index] + memoization_mincost_stairs(index+2)  # cost of step is same as ith-number at cost-array
        
        min_cost[index]= min(one_step,two_step)  # add minimum -cost to min-cost-array at respective indexes  
        return min_cost[index]
    
    #now need to initialize recursive function with intial indexes [either from index-0 or index-1] even from that we're going to return minimum-cost among them 
    return min (memoization_mincost_stairs(0), memoization_mincost_stairs(1))
    
    
#example-usage:
cost = [10,20,30]   # min-cost should be 20 from index-1 and reaching top while climbing 2-step
print(Min_Cost_ClimbingStairs(cost))
```
