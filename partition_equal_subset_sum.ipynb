# partition_equal_subset_sum problem statement [with input and expected ouput example]:

### Given an integer array nums, return true if you can partition the array into two subsets such that the sum of the elements in both subsets is equal or false otherwise.

Example 1:

Input: nums = [1,5,11,5]

Output: true

Explanation: The array can be partitioned as [1, 5, 5] and [11].

Example 2:

Input: nums = [1,2,3,5]

Output: false

Explanation: The array cannot be partitioned into equal sum subsets.
 
- clarifying question :
- Is all numbers in array are +ve ?    : yes all numbers in given array are +ve
- means total_sum of all numbers in array will be +ve in all the scenerio

# Intuition arround the approach :
### problem at hand : Given an integer array nums, we need to determine if it can be partitioned into two subsets with equal sums.
### as firstly we should find the total_sum of numbers in an given array
### If the total sum of the array is Odd, return false since it cannot be split into two equal integer sums.  //this will be major-case to return False.

### if the total sum of numbers is Even then we're all set to partition the array into two subsets with equal sum
### so for forming one of the above two equal sum subset , We should determine the target_sum that will be half of the total_sum of numbers in an given array 

### for example if given array is  nums = [1,5,11,5]  ; so its total_sum = 22 ; so target_sum of both subsets will be : total_sum//2 = 22//2 = 11 ; so target_sum = 11 
### it means if we able to ahcieve target-sum = 11 while building one of the subset then we will return True for given array ; and that will be the solution to problem at hand  

# Recursion-approach (recursion-tree to vaisulaize the above intuition)
### we will go recursively from from index 0 to upto the index of last number in the array 

### for each number on succesive index we have two choices either exclude the numbers or include the number 
### when we exclude the number target_sum reamined same and we will move to the next index 
### when we include the number taeget_sum will be reduced by number and we will move to the next index  

### we will do this inclusion exclusion procedures with each and every number 
### and check whether in inclusuion-branch or exclusion branch we achieved our total-sum or not 

### while going through branches of inclusion and exclusion if our target-sum = 0 then return true //as we found one of the subset with sum equals to target-sum  [base case] 

### while going through branches of inclusion and exclusion if our target-sum = -ve(or less than zero) then return false //as this means while going through branches we can't achieve sum equals to target-sum   [base-case]
### and if index goes beyound length of int-array return False // means no number to include or exclude further to achieve sum equals to target-sum  [base-case]

### recursion-tree(recursion function takes index and curr_sum or sum_so_far as argument)
- note: initialise the recursion fn with index = 0 and target-sum

              /   \

means:  including (1)-> including (5)-> excluding (11)- > including(5)  gives curr_sum equals to target-sum
- means, True : we can partition the given array in to two subset of equal int-sum 

### if we use recursive approach then time complexity will be of O(2^n) //as depth binary tree here 'n' and each node has two branches(excl & incl)   
### and space complexity will be O(n) //dpeth of recursive call-stack

# optimizing the complexity by using memoisation-approach(top-down) to solve the partition_equal_subset_sum problem:

### here we will make the 2-d dp_table with row equals to the length of array and columns from target-sum to zero 
### here for implemeneting the dp_table we're using the hash-table which stores the True value at the corresponding index and curr_sum value from exclusion and inclusion branch (if one of them will be true) 

# now time complexity of memoisation approach will be:  O(n*target-sum) = O(n * T)
# and the space complexity of memoisation appproach will be also : O(n*target-sum) = O(n * T) //'coz we contruct 2-d dp_table of n-rows and T-columns (from target-sum to zero)

```python
#code implementation of memoisation apporach (top-down) for solving the partition_equal_subset problem
def can_partition(nums):
    total_sum = sum(nums)
    
    #Major-case :  If the total sum is odd, it cannot be partitioned into two equal subsets  
    if total_sum % 2 != 0:
        return False
    
    target = total_sum // 2   # calculating tatget-sum by make half of total sum of numbers  in array
    dp = {}   # initialised hashtable /dp_table which stores the True-value at corresponding index and curr_sum if their curr_sum equals to target_sum
    
    def can_partition_recursive(index, current_sum):
        # Base cases
        # base-case-1 : when curr_sum equals to zero means we achieved target-sum through recursively exclusion and inclusion of nums     
        if current_sum == 0:
            return True
        #base-case-2: if index reached last number in array or curr_sum is _ve // means taeget-sum not achieved
        if index >= len(nums) or current_sum < 0:
            return False
        
        # Check dp_table for index and curr_sum if its existed then return value(sub-solution) at corresponding index and curr_sum
        if (index, current_sum) in dp:
            return dp[(index, current_sum)]
        #else caluclate its solutuon by inclusion and exclusion
        # Include the current element ; next-index call and curr_sum reduced by number at current index
        include = can_partition_recursive(index + 1, current_sum - nums[index])
        
        # Exclude the current element ; next-index call and curr_sum will be reamined same
        exclude = can_partition_recursive(index + 1, current_sum)
        
        # Store result in memo  from include or exclude whichever gives True 
        dp[(index, current_sum)] = include or exclude
        return dp[(index, current_sum)]
    
    return can_partition_recursive(0, target)   # initialised fn with index 0 and curr_sum = target-sum


# example-usage :
```
